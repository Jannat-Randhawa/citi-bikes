```python
#Dependencies 
import pandas as pd 
import datetime as dt 
import numpy as np 
```

## April Data

```python
#file 
april18 = "Resources/201804-citibike-tripdata.csv"
april19 = "Resources/201804-citibike-tripdata.csv"
april20 = "Resources/JC-202004-citibike-tripdata.csv"

#load the data in data frames and merge them together. 
april_df18 = pd.read_csv(april18, encoding="ISO-8859-1")
april_df19 = pd.read_csv(april19, encoding="ISO-8859-1")
april_df20 = pd.read_csv(april20, encoding="ISO-8859-1")
```

```python
# Combine all the data together 
april_data = april_df18.append([april_df19, april_df20], ignore_index=True)
april_data.head()
```

```python
# Change columns to sentence form. 
april_data.columns = april_data.columns.str.title()
# Change columns names 
april_data = april_data.rename(columns={
    "Tripduration" : "Trip Duration", 
    "Starttime": "Start Time",
    "Stoptime" : "Stop Time",
    "Bikeid": "Bike ID", 
    "Usertype": "User Type", 
    "Gender" : "Gender Attr"
})
april_data.head()
```

```python
# Convert gender value from binary to string 
def findGender(gender):
    if gender == 0:
        return "Unknown"
    elif gender ==1:
        return "Male"
    else:
        return "Female"
april_data["Gender"] = april_data["Gender Attr"].apply(findGender)
```

```python
# Drop the gender attr values from the data frame
april_data = april_data.drop(['Gender Attr'], axis=1)
```

```python
april_data.head()
```
