```python
#Dependencies 
import pandas as pd 
import datetime as dt 
import numpy as np 
```

## June Data 

```python
#file 
june13 = "Resources/201306-citibike-tripdata.csv"
june18 = "Resources/201806-citibike-tripdata.csv"
june19 = "Resources/201906-citibike-tripdata.csv"
june20 = "Resources/JC-202006-citibike-tripdata.csv"

#load the data in data frames and merge them together. 
june_df13 = pd.read_csv(june13, encoding="ISO-8859-1")
june_df18 = pd.read_csv(june18, encoding="ISO-8859-1")
june_df19 = pd.read_csv(june19, encoding="ISO-8859-1")
june_df20 = pd.read_csv(june20, encoding="ISO-8859-1")
```

```python
# Combine all the data together 
june_data = june_df13.append([june_df18, june_df19, june_df20], ignore_index=True)
june_data.head()
```

```python
# Change columns to sentence form.
june_data.columns = june_data.columns.str.title()

# Change columns names 
june_data = june_data.rename(columns={
    "Tripduration" : "Trip Duration", 
    "Starttime": "Start Time",
    "Stoptime" : "Stop Time",
    "Bikeid": "Bike ID", 
    "Usertype": "User Type",
    "Gender" : "Gender Attr"
})
june_data.head()
```

```python
# Convert gender value from binary to string 
def findGender(gender):
    if gender == 0:
        return "Unknown"
    elif gender ==1:
        return "Male"
    else:
        return "Female"
june_data["Gender"] = june_data["Gender Attr"].apply(findGender)

# Drop the gender attr values from the data frame
june_data = june_data.drop(['Gender Attr'], axis=1)
june_data.head()
```

```python
# #Save to csv 
june_data.to_csv("combined_june.csv", index=False, header=True)
```
